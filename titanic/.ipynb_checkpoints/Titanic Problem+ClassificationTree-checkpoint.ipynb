# =======Building the Classifier from Scrach=======

### TRAIN-TEST-SPLIT

```python
# Helper function for splitting the data
def train_test_split(df, test_size):
    if(isinstance(test_size, float)):
        test_size = round(test_size*len(df))
    index_list = df.index.tolist()
    random.seed(1)
    test_indices = random.sample(population=index_list, k=test_size)
    test_df = df.loc[test_indices]
    train_df = df.drop(test_indices, axis=0)
    return train_df, test_df
```

```python
train_df,test_df = train_test_split(df, 0.3);
```

## HELPER FUNCTIONS

### IS PURE?

```python
def check_purity(data):
    label_col = data[:, -1]
    unique_classes = np.unique(label_col)
    if(len(unique_classes)==1):
        return True
    else:
        return False
```

```python
d = train_df[train_df["Survived"]==1].values
check_purity(d)
```

### CLASSIFY

```python
def classify_data(data):
    label_col = data[:, -1]
    unique_labels, unique_label_counts = np.unique(label_col, return_counts=True)
    majorirty_index = unique_label_counts.argmax()
    return unique_labels[majorirty_index]
```

### POTENTIAL SPLITS

```python
def get_potential_splits(data):
    _,n_columns = data.shape
    potential_splits = {}
    for col_index in range(n_columns):
        unique_values = np.unique(data[:,col_index])
        potential_splits[col_index]=[]
        for index in range(1,len(unique_values)):
            previous_value = unique_values[index-1]
            current_value = unique_values[index]
            potential_split = (previous_value + current_value) / 2
            potential_splits[col_index].append(potential_split)
        
    return potnetial_splits
```
